```python
%matplotlib inline

import pandas as pd

# these will be useful later
import numpy  as np
from matplotlib import pyplot as plt
```

```python
#Homicide Rates in 2013
crime=pd.read_csv("mun.csv", index_col = "id", parse_dates = ["id"], encoding='iso-8859-1')
crime=crime.drop('SUBTIPO', axis=1)
```

```python
hom13= crime[(crime['MODALIDAD']=='HOMICIDIOS') & (crime['TIPO'] == 'DOLOSOS' )]
hom13= hom13.groupby([hom13.index,'year','ENTIDAD', 'MUNICIPIO']).sum().reset_index()
hom13=hom13.set_index(['id'])
hom13=hom13[hom13.year == 2013]
hom13=hom13.rename(columns = {'total':'homicidios'})
```

```python
pop=pd.read_csv("pop.csv", index_col = "id", parse_dates = ["id"], encoding='iso-8859-1')
pop=pop.drop('hom', axis=1)
pop=pop.drop('muj', axis=1)
```

```python
tot13 = pd.concat([hom13, pop], axis=1)
tot13= tot13.drop('nom', axis=1)
tot13['hom_rate'] = (tot13['homicidios']/tot13['pop'])*100000
tot13 = tot13.drop("MUNICIPIO", axis=1)
tot13 = tot13.rename(columns = {'hom_rate': 'Homicide Rate in 2013',
                               'homicidios':'Homicides in 2013'})
```

```python
#Homicide Rates in 2014
crime=pd.read_csv("mun.csv", index_col = "id", parse_dates = ["id"], encoding='iso-8859-1')
crime=crime.drop('SUBTIPO', axis=1)
hom14= crime[(crime['MODALIDAD']=='HOMICIDIOS') & (crime['TIPO'] == 'DOLOSOS' )]
hom14= hom14.groupby([hom14.index,'year','ENTIDAD', 'MUNICIPIO']).sum().reset_index()
hom14=hom14.set_index(['id'])
hom14=hom14[hom14.year == 2014]
hom14=hom14.rename(columns = {'total':'homicidios'})
pop=pd.read_csv("pop.csv", index_col = "id", parse_dates = ["id"], encoding='iso-8859-1')
pop=pop.drop('hom', axis=1)
pop=pop.drop('muj', axis=1)
tot14 = pd.concat([hom14, pop], axis=1)
tot14= tot14.drop('nom', axis=1)
tot14['hom_rate'] = (tot14['homicidios']/tot14['pop'])*100000
tot14 = tot14.rename(columns = {'hom_rate': 'Homicide Rate in 2014',
                               'homicidios':'Homicides in 2014'})

```

```python
#3x1 Projects Granted in 2013
x31_13 = pd.read_csv("3x1_2013utf8.csv", index_col = "id", usecols= {"id","year", "suma"}, 
                     parse_dates = ["id"])
x31_13 = x31_13.groupby([x31_13.index,'year']).sum()
x31_13 = x31_13.rename(columns = {'suma': '3x1 Projects Granted in 2013'})
x31_13 = x31_13.reset_index()
x31_13 = x31_13.set_index(['id'])
```

```python
#3x1 Projects Granted in 2014
x31_14 = pd.read_csv("3x1_2014utf8.csv", index_col = "id", usecols= {"id","year", "suma"}, 
                     parse_dates = ["id"])
x31_14 = x31_14.groupby([x31_14.index,'year']).sum()
x31_14 = x31_14.rename(columns = {'suma': '3x1 Projects Granted in 2014'})
x31_14 = x31_14.reset_index()
x31_14 = x31_14.set_index(['id'])
x31_14.head(10)
```

```python
#Merging homicide rates and 3x1 projects granted in 2013
hom_3x1_13 = pd.merge(tot13, x31_13, left_index=True, right_index=True, how='outer')
hom_3x1_13.head(10)
```

```python
#Merging homicide rates and 3x1 projects granted in 2014
hom_3x1_14 = pd.merge(tot14, x31_14, left_index=True, right_index=True, how='outer')
hom_3x1_14.head(10)
```

```python
#Dropping repeated columns in 2013
hom_3x1_13 = hom_3x1_13.drop('year_x', axis=1)
hom_3x1_13 = hom_3x1_13.drop('year_y', axis=1)
hom_3x1_13.head(10)
```

```python
#Dropping repeated columns in 2014
hom_3x1_14 = hom_3x1_14.drop('year_x', axis=1)
hom_3x1_14 = hom_3x1_14.drop('year_y', axis=1)
hom_3x1_14.head(10)
```

```python
#Final Merge!!!!
hom_3x1_all = pd.merge(hom_3x1_13, hom_3x1_14, left_index=True, right_index=True, how='outer')
hom_3x1_all.head(10)
```

```python
# Dropping repeated columns
hom_3x1_all = hom_3x1_all.drop('ENTIDAD_x', axis=1)
hom_3x1_all = hom_3x1_all.drop('pop_x', axis=1)
hom_3x1_all = hom_3x1_all.drop('ENTIDAD_y', axis=1)
hom_3x1_all = hom_3x1_all.drop('pop_y', axis=1)
hom_3x1_all = hom_3x1_all.drop('MUNICIPIO', axis=1)
hom_3x1_all.head(10)
```

```python
hom_3x1_all.to_csv('master.csv')
```
